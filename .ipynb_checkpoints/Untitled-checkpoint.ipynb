```python
import numpy as np
from sklearn.neighbors import NearestNeighbors
import faiss
import time
import matplotlib.pyplot as plt

# Generate a synthetic dataset with 100,000 points and 128 dimensions
np.random.seed(42)
n_samples = 10000
n_features = 300
data = np.random.random((n_samples, n_features)).astype('float32')

# Random query point
query = np.random.random((1, n_features)).astype('float32')

# Number of nearest neighbors to search for
k = 5

# --- Sklearn Nearest Neighbors ---
# Fit the sklearn NearestNeighbors model
start_time = time.time()
nn_sklearn = NearestNeighbors(n_neighbors=k, algorithm='brute', metric='euclidean')
nn_sklearn.fit(data)
distances_sklearn, indices_sklearn = nn_sklearn.kneighbors(query)
end_time = time.time()
time_sklearn = end_time - start_time
print(f"Scikit-learn time: {time_sklearn:.4f} seconds")

# --- Faiss Nearest Neighbors ---
# Create index for Faiss
index = faiss.IndexFlatL2(n_features)  # L2 distance index
start_time = time.time()
index.add(data)  # Add data to the index
distances_faiss, indices_faiss = index.search(query, k)  # Perform search
end_time = time.time()
time_faiss = end_time - start_time
print(f"Faiss time: {time_faiss:.4f} seconds")

# Comparison of the results
print("\nNearest neighbors (Scikit-learn):", indices_sklearn)
print("Nearest neighbors (Faiss):", indices_faiss)

# Plotting the results
labels = ['Scikit-learn', 'Faiss']
times = [time_sklearn, time_faiss]

plt.bar(labels, times, color=['blue', 'orange'])
plt.ylabel('Time (seconds)')
plt.title('Nearest Neighbor Search Time Comparison')
plt.ylim(0, max(times) + 1)  # Adjust the y-axis for better visualization
plt.grid(axis='y')

# Show the plot
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
import faiss
import time
import matplotlib.pyplot as plt

# Generate a synthetic dataset with 100,000 points and 128 dimensions
np.random.seed(42)
n_samples = 100000
n_features = 128
data = np.random.random((n_samples, n_features)).astype('float32')

# Number of clusters for K-means
n_clusters = 10

# --- Sklearn K-means ---
start_time = time.time()
kmeans_sklearn = KMeans(n_clusters=n_clusters, random_state=42)
kmeans_sklearn.fit(data)
end_time = time.time()
time_sklearn = end_time - start_time
print(f"Scikit-learn K-means time: {time_sklearn:.4f} seconds")

# --- Faiss K-means ---
# Faiss K-means uses Euclidean distance, and the 'niter' argument specifies the number of iterations
kmeans_faiss = faiss.Kmeans(d=n_features, k=n_clusters, niter=20, verbose=False)

start_time = time.time()
kmeans_faiss.train(data)
end_time = time.time()
time_faiss = end_time - start_time
print(f"Faiss K-means time: {time_faiss:.4f} seconds")

# Plotting the results
labels = ['Scikit-learn', 'Faiss']
times = [time_sklearn, time_faiss]

plt.bar(labels, times, color=['blue', 'orange'])
plt.ylabel('Time (seconds)')
plt.title('K-means Clustering Time Comparison')
plt.ylim(0, max(times) + 1)  # Adjust the y-axis for better visualization
plt.grid(axis='y')

# Show the plot
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import faiss
import time
import matplotlib.pyplot as plt

# Generate a synthetic dataset with 100,000 points and 128 dimensions
np.random.seed(42)
n_samples = 100000
n_features = 128
data = np.random.random((n_samples, n_features)).astype('float32')

# Number of clusters for K-means
n_clusters = 10

# --- Sklearn K-means ---
start_time = time.time()
kmeans_sklearn = KMeans(n_clusters=n_clusters, random_state=42)
kmeans_sklearn.fit(data)
end_time = time.time()
time_sklearn = end_time - start_time
print(f"Scikit-learn K-means time: {time_sklearn:.4f} seconds")

# --- Faiss K-means ---
# Faiss K-means uses Euclidean distance, and the 'niter' argument specifies the number of iterations
kmeans_faiss = faiss.Kmeans(d=n_features, k=n_clusters, niter=20, verbose=False)

start_time = time.time()
kmeans_faiss.train(data)
end_time = time.time()
time_faiss = end_time - start_time
print(f"Faiss K-means time: {time_faiss:.4f} seconds")

# --- Dimensionality Reduction (PCA for visualization) ---
# Use PCA to reduce the dataset to 2 dimensions for visualization purposes
pca = PCA(n_components=2)
data_2d = pca.fit_transform(data)

# Get cluster labels for the data points from scikit-learn KMeans
labels_sklearn = kmeans_sklearn.labels_

# Get cluster labels for the data points from Faiss KMeans
_, labels_faiss = kmeans_faiss.index.search(data, 1)  # Faiss uses its own method to assign points to clusters

# Plotting the results for Scikit-learn Clustering
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(data_2d[:, 0], data_2d[:, 1], c=labels_sklearn, cmap='viridis', s=1)
plt.title('Scikit-learn K-means Clustering')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')

# Plotting the results for Faiss Clustering
plt.subplot(1, 2, 2)
plt.scatter(data_2d[:, 0], data_2d[:, 1], c=labels_faiss.flatten(), cmap='plasma', s=1)
plt.title('Faiss K-means Clustering')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')

# Show the plot
plt.tight_layout()
plt.show()
```
